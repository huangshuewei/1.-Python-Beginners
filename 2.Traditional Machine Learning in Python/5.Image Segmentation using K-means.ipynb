```python
import numpy as np
import cv2

img = cv2.imread('image/BSE_Google.jpg')

print(img.shape)
```

```python
img2 = img.reshape((-1,3))

print(img2.shape)
```

```python
img2 = np.float32(img2)

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,10, 1.)

# Clusters
k=4
attempts = 10

ret, label, centre = cv2.kmeans(img2, k, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)

centre = np.uint8(centre)
res = centre[label.flatten()]
res2 = res.reshape((img.shape))

```

```python
cv2.imwrite('segmented.jpg',res2)
```

```python
# img2 = np.float32(img2)

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,10, 1.)

# Clusters
k=5
attempts = 10

ret, label, centre = cv2.kmeans(img2, k, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)

centre = np.uint8(centre)
res = centre[label.flatten()]
res2_k5 = res.reshape((img.shape))

cv2.imwrite('segmented_K5.jpg',res2_k5)
```

```python
from matplotlib import pyplot as plt

plt.figure()
plt.imshow(img)
plt.figure()
plt.imshow(res2)
plt.figure()
plt.imshow(res2_k5)
```
