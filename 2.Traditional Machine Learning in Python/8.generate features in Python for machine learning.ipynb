```python
import cv2
from skimage.filters.rank import entropy
from skimage.morphology import disk
from matplotlib import pyplot as plt

img = cv2.imread('image/scratch.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

entropy_img = entropy(img, disk(1))

plt.figure()
plt.imshow(img, cmap='gray')
plt.figure()
plt.imshow(entropy_img, cmap='gray')
```

```python
img = cv2.imread('image/Yeast_Cells.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

entropy_img = entropy(img, disk(1))

plt.figure()
plt.imshow(img, cmap='gray')
plt.figure()
plt.imshow(entropy_img, cmap='gray')
## entropy filter is not suitable to this case
```

```python
from scipy import ndimage as nd

gaussian_img = nd.gaussian_filter(img, sigma=3)

plt.figure()
plt.imshow(img, cmap='gray')
plt.figure()
plt.imshow(gaussian_img, cmap='gray')
```

```python
from skimage.filters import sobel

sobel_img = sobel(img)

plt.figure()
plt.imshow(img, cmap='gray')
plt.figure()
plt.imshow(sobel_img, cmap='gray')
```

```python
img2 = img.reshape(-1)

import pandas as pd

df = pd.DataFrame()
df['Original_pixel_values'] = img2

entropy1 = entropy_img.reshape(-1)

df['Entrop'] = entropy1

gaussian1 = gaussian_img.reshape(-1)
df['Gaussian'] = gaussian1

sobel1 = sobel_img.reshape(-1)
df['Sobel'] = sobel1

print(df.head(25))
```
