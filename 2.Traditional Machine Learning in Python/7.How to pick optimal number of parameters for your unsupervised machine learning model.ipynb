```python
import numpy as np
import cv2
from sklearn.mixture import GaussianMixture as GMM

img = cv2.imread('image/Alloy.jpg')
img2 = img.reshape((-1,3))

n_components = np.arange(1,10)

gmm_model = [GMM(n, covariance_type='tied').fit(img2) for n in n_components]

from matplotlib import pyplot as plt
plt.plot(n_components, [m.bic(img2) for m in gmm_model], label='BIC')
plt.xlabel('n_components')
```

```python
# Pick n=3
gmm_model = GMM(n_components=3, covariance_type='tied').fit(img2)
gmm_labels = gmm_model.predict(img2)

original_shape = img.shape
segmented = gmm_labels.reshape(original_shape[0], original_shape[1])

plt.figure()
plt.imshow(segmented, cmap='gray')
```
