```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

ksize = 5
sigma = 3
theta = 1*np.pi/4
lamda = 1*np.pi/4
gamma = 0.5
phi = 0

kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)

plt.imshow(kernel)
```

```python
ksize = 5
sigma = 3
theta = 1*np.pi/4
lamda = 1*np.pi/4
gamma = 0.5
phi = 0

kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)

img = cv2.imread('image/BSE_Google.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
fimg = cv2.filter2D(img, cv2.CV_8UC3, kernel)
kernel_resize = cv2.resize(kernel, (400, 400))

plt.figure()
plt.imshow(img, "gray")
plt.figure()
plt.imshow(kernel_resize, "gray")
plt.figure()
plt.imshow(fimg, "gray")
```
