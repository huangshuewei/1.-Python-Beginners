```python
def my_function():
    print("Hello from inside the function")

def my_function_name(my_name):
    print("My name is:", my_name)

my_function()
my_function_name("Wayne")
```

```python
def my_function_country(country="Taiwan"):
    print("My country is:", country)

my_function_country()
my_function_country("United Kingdom")
```

```python
def my_function_food(food):
    for x in food:
        print(x)

fruits = ['apple', 'pear', 'banana']
my_function_food(fruits)
```

```python
def my_function_five(x):
    return 5*x

print(my_function_five(3))
```

```python
def F(C):
    return print('%d degrees Centigrade is equal to %d Fahrenheit' %(C, (9./5)*C+32))

F(-40)
```

```python
def F(C):
    for i in C:
        F = (9./5)*i+32
        print('%d degrees Centigrade is equal to %d Fahrenheit' %(i, F))

Centi = [-40,-20,-10,0,10,20,40]
F(Centi)
```

```python
from scipy import misc, ndimage
from matplotlib.pyplot import imread, imsave

def rotated(img):
    rotated_image = ndimage.rotate(img, 45)
    imsave('rotated.jpg', rotated_image/255)
    return rotated_image/255

def blurred(img):
    blurred_image = ndimage.gaussian_filter(img, sigma=5)
    imsave('blurred.jpg', blurred_image/255)
    return blurred_image/255

def denoised(img):
    denoised_image = ndimage.median_filter(img, 3)
    imsave('denoised.jpg', denoised_image/255)
    return denoised_image/255


img1 = imread('test_image.jpg')
img2 = rotated(img1)
img3 = blurred(img1)
img4 = denoised(img1)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(img2)
plt.figure()
plt.imshow(img3)
plt.figure()
plt.imshow(img4)
```
