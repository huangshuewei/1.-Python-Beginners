```python
"""
Ridge filters can be used to detect ridge-like structures, such as neurites 1, tubes 2, vessels 3, wrinkles 4 or rivers.
"""
```

```python
from skimage import io, filters, feature
import matplotlib.pyplot as plt
from skimage.color import rgb2gray

from skimage.filters import meijering, sato, frangi, hessian

img = io.imread("images/leaf.jpg")
img = rgb2gray(img)

meijering_img = meijering(img)
sato_img = sato(img)
frangi_img = frangi(img)
hessian_img = hessian(img)
```

```python
plt.figure()
plt.imshow(img, cmap='gray')
plt.title('Input image')

plt.figure()
plt.imshow(meijering_img, cmap='gray')
plt.title('meijering image')

plt.figure()
plt.imshow(sato_img, cmap='gray')
plt.title('sato image')

plt.figure()
plt.imshow(frangi_img, cmap='gray')
plt.title('frangi image')

plt.figure()
plt.imshow(hessian_img, cmap='gray')
plt.title('hessian image')
```
